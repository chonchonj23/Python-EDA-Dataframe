```python
# Scatter Plot, color = "pink"

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(18, 10))
df_housing.plot(ax=axes[0,0], x="AveOccup", y="MedHouseVal", kind="scatter", color = "pink")
df_housing.plot(ax=axes[0,1], x="HouseAge", y="MedHouseVal", kind="scatter", color = "pink")
df_housing.plot(ax=axes[0,2], x="Population", y="MedHouseVal", kind="scatter", color = "pink")

df_housing.plot(ax=axes[1,0], x="AveOccup", y="MedInc", kind="scatter", color = "pink")
df_housing.plot(ax=axes[1,1], x="HouseAge", y="MedInc", kind="scatter", color = "pink")
df_housing.plot(ax=axes[1,2], x="Population", y="MedInc", kind="scatter", color = "pink")
```
